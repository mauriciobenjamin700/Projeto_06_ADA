```python
dados = df.unit_price.unique()
```

```python
for dado in dados:
    print(dado)
```

```python


valores = []


for dado in dados:
    #formatando nossos dados para númericos
    valores.append(float(dado.replace(",",".").split(" ")[0]))
    
print(valores)
```

```python
print(f"Mais caro: {max(valores)}\nMais barato: {min(valores)}")

print(f"Mediana: {np.median(valores)}")

print(f"\nMédia: {np.mean(valores)}\nDesvio Padrão: {np.std(valores)}\nVariancia: {np.var(valores)}")
```
